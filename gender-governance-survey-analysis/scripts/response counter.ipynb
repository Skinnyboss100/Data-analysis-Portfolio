```python
import pandas as pd

# Step 1: Load the file
df = pd.read_excel(r"C:\Users\NICHOLAS\Desktop\prof analysis\sort5split.xlsx")

# Step 2: Identify all response columns (response1, response2, ...)
response_cols = [col for col in df.columns if col.lower().startswith("response")]

# Step 3: Convert wide format to long format
# Each response becomes one row (one vote)
long_df = (
    df[response_cols]
    .melt(value_name="response")
    .dropna(subset=["response"])
)

# Step 4: Clean whitespace
long_df["response"] = long_df["response"].astype(str).str.strip()

# Step 5: Count each response
response_counts = (
    long_df["response"]
    .value_counts()
    .reset_index()
    .rename(columns={"index": "Response", "response": "Count"})
)

# Step 6: View results
print(response_counts)

response_counts["Percentage"] = (
    response_counts["Count"] / response_counts["Count"].sum()
) * 100



response_counts.to_excel(r"C:\Users\NICHOLAS\Desktop\prof analysis\sort5count.xlsx", index=False)
```
