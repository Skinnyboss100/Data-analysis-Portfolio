```python
import pandas as pd
import re

df = pd.read_excel(r"C:\Users\NICHOLAS\Desktop\prof analysis\sort1.xlsx")
```

```python
column_name = 'column'
split_pattern = r',\s*(?![^()]*\))'
responses = df[column_name].str.split(split_pattern, expand=True)
responses.columns = [f"response{i+1}" for i in range(responses.shape[1])]

df = pd.concat([df, responses], axis=1)
df[responses.columns] = df[responses.columns].apply(lambda x: x.str.strip())
```
