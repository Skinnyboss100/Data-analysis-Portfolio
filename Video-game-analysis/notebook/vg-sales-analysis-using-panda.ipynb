```python
#I normalised the column names by changing the column names to lower case

df.columns = df.columns.str.strip().str.lower()
```

```python
df
```

```python
# I made all the numeric sales are in floats using the to_numeric function

for row_data in ['na_sales','eu_sales','jp_sales','global_sales','other_sales']:
    df[row_data] = pd.to_numeric(df[row_data],errors='coerce').fillna(0.0)
```

```python
df.head()
```

```python
# I checked the duplicates using excel, so i will check for outliers

df.nlargest(10,'global_sales')
```

```python
#Exploratory Data Analysis using groupby

#Firstly, top 10 genres by total global sales

genre_sales = df.groupby('genre',as_index=False)['global_sales'].sum().sort_values('global_sales',ascending=False)
```

```python
genre_sales
```

```python
#Secondly, top 10 platforms by total global sales

platform = df.groupby('platform',as_index=False)['global_sales'].sum().sort_values('global_sales',ascending=False).head(10)
```

```python
platform
```

```python
#Thirdly, top 10 publishers by total global sales

publisher = df.groupby('publisher',as_index=False)['global_sales'].sum().sort_values('global_sales',ascending=False).head(10)
```

```python
publisher
```

```python
#Lastly, top 10 sales overtime

yearly = df.groupby('year',as_index=False)['global_sales'].sum().dropna().sort_values('year').head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yearly_sales = df.groupby('year',as_index=False)['global_sales'].sum()

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_sales,x='year',y='global_sales')

plt.title('Global Video game Sales Overtime')
plt.xlabel('year')
plt.ylabel('Total Global Sales')
plt.grid(True)
plt.show()
```

```python
#Bar chart showing top genre by global sales
#I used bar charts clearly show winners vs weaker genres...Required for comparative analysis

genre_sales = df.groupby('genre',as_index=False)['global_sales'].sum()
genre_sales = genre_sales.sort_values('global_sales',ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=genre_sales,x='genre',y='global_sales')

plt.title('Top genres by global sales')
plt.xlabel('Genre')
plt.ylabel('Global sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Boxplot...I used the boxplot to see which genres consistently performed over the years

plt.figure(figsize=(14,7))
sns.boxplot(data=df,x='genre',y='global_sales')

plt.title('Distribution of global sales by genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Scatter plot to detect major blockbuster year and to compare regional strengths,the blockbuster years
#-will appear as large bubbles

plt.figure(figsize=(10,7))

sns.scatterplot(
    data=df,
    x='year',
    y='global_sales',
    size= 'na_sales',
    hue='eu_sales',
    sizes=(20,300),
    alpha=0.6
)

plt.title('Year vs Global Sales(sized by NA_sales and colored by EU sales)')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.legend(title='EU sales/NA sales(size)',bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```
